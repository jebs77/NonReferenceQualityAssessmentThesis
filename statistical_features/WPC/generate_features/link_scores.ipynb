```python
import pandas as pd
import numpy as np

vsense = pd.read_csv("CSV/vsense.csv")
scores = pd.read_csv("CSV/mos_std.csv")

print(len(scores))
print(len(vsense))
vsense["score"] = 0

vsense.head()
```

```python
for i in range(len(scores)):
    for j in range(300):
        vsense["score"][i*300+j] = np.random.normal(scores["mos"][i], (scores["std"][i])/10)

vsense.head()

vsense.to_csv("with_score.csv")
```
