```python
import csv

filename = "ratings.csv"
data = []
fields = []

with open(filename, 'r') as csv_file:
    # initialise csv_reader
    csvreader = csv.reader(csv_file)

    # initialise data
    fields = ["name", "transformation_type", "transformation", "object", "background_colour", "rating"]

    for row in csvreader:
        row[-1] = int(row[-1])
        data.append(row)

    print("Total of %d rows" % (csvreader.line_num))

mos_data = []
for i in data:
    found = False

    for j in mos_data:
        if i[1:-1] == j[0:-1]:
            j[-1] += i[-1]
            found = True
            break
    if not found:
        mos_data.append(i[1:])

for i in mos_data:
    i[-1] /= 100

print(mos_data)
print(len(mos_data))

fields_write = ["transformation_type", "transformation", "object", "background_colour", "rating"]

filename = "ratings_mos.csv"

# Add an empty line to the CSV file
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(fields_write)  # Writing header row
    csvwriter.writerows(mos_data)  # Writing each row of aggregated data
```
